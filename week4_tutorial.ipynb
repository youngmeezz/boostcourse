```python
df = df_raw.melt(id_vars=["국가(대륙)별","상품군별","판매유형별"], var_name="기간", value_name="백만원")

df.shape
```

# 5 데이터 전처리

# 5.1 기간에서 연도를 분리하기

```python
int("2022 3/4 p)".split()[0])
```

```python
df["연도"] = df["기간"].map(lambda x : int(x.split()[0]))
df.head()
```

```python
df["기간"].map
```

```python
"2022 3/4 p)".split()[1]
```

```python
"2022 3/4 p)".split()[1].split("/")[0]
```

```python
# 기간에서 분기만 분리하기
df["분기"] = df["기간"].map(lambda x : int(x.plit()[1].split("/")[0]))
df.head()
```

```python
df["백만원"] = df["백만원"].replace("-", pd.np.nan).astype(float)
df["백만원"]
df.head()
```

# 5.3 필요없는 데이터 제거하기

```python
df.info()
```

```python
# 합계 데이터는 따로 구할 수 있기 떄문에 전체 데이터에서 제거합니다.
df = df[(df["국가(대륙)별"] != "합계") & (df["상품군별"] != "합계")].copy()
df.info()
```

```python
# 결측치 보기
df.isnull().sum()
```

# 6 K-beauty 시각화

# 6.1 전체 상품군 판매액

```python
df_total = df[df["판매유형별"] == "계"].copy()
df_total.head()
```

```python
#연도, 판매액 lineplot그리기
sns.lineplot(data=df_total, x="연도", y="백만원")
```

```python
#연도, 판매액 lineplot으로 그리고 상품군별로 다른 색상으로 표시하기
sns.lineplot(data=df_total, x="연도", y="백만원", hue="상품군별")
plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
```

```python
# 위에 그린 그래프를 자세히 보기 위해 서브플롯으로 표시하기
sns.relplot(data=df_total, x="연도", y="백만원", hue="상품군별", kind="line", col="상품군별", col_wrap=4)
```

```python
#isin을 사용해 화장품만 제외를 하고 df_sub이라는 변수에 담습니다.
df_sub = df_total[~df_total["상품군별"].isin(["화장품", "의류 및 패션관련 상품"])].copy()
```

```python
# 연도별 판매액을 상품군별로 relplot을 활용해 서브플롯으로 그려보기

sns.relplot(data=df_sub, x="연도", y="백만원", hue="상품군별", col="상품군별", col_wrap=4,kind="line")
```

# 6.2 화장품의 온라인쇼핑 해외직접판매액

```python
# df_cosmetic 이라는 변수에 상품군별이 화장품인 데이터만 가져오기

df_cosmetic = df_total[df_total["상품군별"] == "화장품"].copy()
df_cosmetic["상품군별"].unique()
```

```python
# 연도와 판매액을 lineplot으로 그리고 분기별로 다른 색상으로 표현하기
plt.figure(figsize=(15,4))
sns.lineplot(data=df_cosmetic, x="연도", y="백만원", hue="분기")
```

```python
df_cosmetic.head()
```

```python
# 화장품 판매액에 대한 기간별 금액 데이터 시각화
plt.figure(figsize=(15,4))
plt.xticks(rotation=30)
sns.lineplot(data=df_cosmetic, x="기간", y="백만원")
```

```python
# 화장품 판매액에 대한 기간별 금액 데이터 시각화 하고 "국가(대륙)별"로 다른 색상으로 표시하기
plt.figure(figsize=(15,4))
plt.xticks(rotation=30)
sns.lineplot(data=df_cosmetic, x="기간", y="백만원", hue="국가(대륙)별")
```

```python
# 화장품 판매액에 대한 기간별 금액 데이터 시각화 하고 "국가(대륙)별"로 다른 색상으로 표시하기
plt.figure(figsize=(15,4))
plt.xticks(rotation=30)
sns.lineplot(data=df_cosmetic[df_cosmetic["국가(대륙)별"] !="중국"], x="기간", y="백만원", hue="국가(대륙)별")
```

```python
#화장품 판매액에 대한 기간별 금액 데이터 시각화 "판매유형별"로 다른색상
plt.figure(figsize=(15,4))
plt.xticks(rotation=30)
df_sub = df[df["판매유형별"] != "계"].copy()
sns.lineplot(data=df_sub, x="기간", y="백만원", hue="판매유형별")
```

```python
#화장품 판매액에 대한 기간별 금액 데이터 시각화 "판매유형별"로 다른색상
plt.figure(figsize=(15,4))
plt.xticks(rotation=30)
df_sub = df[(df["판매유형별"] != "계") & (df["판매유형별"]) != "면세점"].copy()
sns.lineplot(data=df_sub, x="기간", y="백만원", hue="판매유형별", ci=None)
```

# 6.3 의류 및 패션관련 상품 온라인 쇼핑 해외직접판매액

```python
# df_fasion이라는 변수에 의류 데이터만 가져와 따로 담기

df_fashion = df[(df["상품군별"] == "의류 및 패션관련 상품") & (df["판매유형별"] == "계")].copy()
df_fashion.head()
```

```python
# "의류 및 패션관련 상품" 판매액에 대한 기간별 금액 데이터 시각화 하고
# "국가(대륙)별" 로 다른 색상으로 표시하기
plt.figure(figsize=(15,4))
plt.xticks(rotation=30)
sns.lineplot(data=df_fashion, x="기간", y="백만원", hue=" 국가(대륙)별")
```

```python
# "의류 및 패션관련 상품" 판매액에 대한 기간별 금액 데이터 시각화하고
#"판매유형별"로 다른 색상으로 표시하기

df_fashion2 = df[(df["상품군별"] == "의류 및 패션관련 상품") & (df["판매유형별"] != "계")].copy()

plt.figure(figsize=(15,4))
plt.xticks(rotation=30)
sns.lineplot(data=df_fashion2, x="기간", y="백만원", hue="판매유형별", ci=None)
```

# 6.4 데이터 집계하기 

```python
df_fashion.head(2)
```

```python
df_fashion["판매유형별"].value_counts()
```

```python
# 피벗테이블로 "국가(대륙)별", "연도"별 합계 금액을 표 형태로 구하기

df_fashion.pivot_table(index="국가(대륙)별", columns="연도", values="백만원", aggfunc="sum")
```

```python
# 피벗테이블로 "국가(대륙)별", "연도"별 합계 금액을 표 형태로 구하기

result = df_fashion.pivot_table(index="국가(대륙)별", columns="연도", values="백만원", aggfunc="mean")

result
```

# 6.5 연산결과를 시각적으로 보기


```python
# 피벗테이블로 구한 결과를 값의 많고 적음에 따라 시각적으로 표현하기

plt.figure(figsize=(10,6))
sns.heatmap(result, cmap="Blues_r", annot=True, fmt=".0f")
```
